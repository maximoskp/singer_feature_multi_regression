```python
import pandas as pd
import os
```

```python
os.makedirs('figs/results', exist_ok=True)
```

```python
baseline = pd.read_csv('results/baseline_frozen_hubert.csv')
nofusion = pd.read_csv('results/nofusion_frozen_hubert.csv')
early = pd.read_csv('results/early_frozen_hubert.csv')
late = pd.read_csv('results/late_frozen_hubert.csv')
```

```python
baseline.head()
```

```python
ax = baseline.plot(
    x = 'epoch',
    y = ['train_loss', 'val_loss'],
    title='baseline loss', xlabel='epoch'
)
fig = ax.get_figure()
fig.savefig('figs/results/' + 'baseline_loss.png', dpi=500)
ax = baseline.plot(
    x = 'epoch',
    y = ['train_acc', 'val_acc'],
    title='baseline accuracy', xlabel='epoch'
)
fig = ax.get_figure()
fig.savefig('figs/results/' + 'baseline_acc.png', dpi=500)
```

```python
ax = nofusion.plot(
    x = 'epoch',
    y = ['train_loss', 'val_loss'],
    title='no fusion loss', xlabel='epoch'
)
fig = ax.get_figure()
fig.savefig('figs/results/' + 'nofusion_loss.png', dpi=500)
ax = nofusion.plot(
    x = 'epoch',
    y = ['train_acc', 'val_acc'],
    title='no fusion accuracy', xlabel='epoch'
)
fig = ax.get_figure()
fig.savefig('figs/results/' + 'nofusion_acc.png', dpi=500)
```

```python
ax = early.plot(
    x = 'epoch',
    y = ['train_loss', 'val_loss'],
    title='early loss', xlabel='epoch'
)
fig = ax.get_figure()
fig.savefig('figs/results/' + 'early_loss.png', dpi=500)
ax = early.plot(
    x = 'epoch',
    y = ['train_acc', 'val_acc'],
    title='early fusion accuracy', xlabel='epoch'
)
fig = ax.get_figure()
fig.savefig('figs/results/' + 'early_acc.png', dpi=500)
```

```python
ax = late.plot(
    x = 'epoch',
    y = ['train_loss', 'val_loss'],
    title='late loss', xlabel='epoch'
)
fig = ax.get_figure()
fig.savefig('figs/results/' + 'late_loss.png', dpi=500)
ax = late.plot(
    x = 'epoch',
    y = ['train_acc', 'val_acc'],
    title='late fusion accuracy', xlabel='epoch'
)
fig = ax.get_figure()
fig.savefig('figs/results/' + 'late_acc.png', dpi=500)
```
